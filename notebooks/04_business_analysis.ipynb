# Muzica Business Analysis

This notebook contains SQL queries used to analyze the Gold layer tables created from the music streaming pipeline.

### Top Artists
Which artists are streamed the most?

```python
%sql
SELECT *
FROM gold_top_artists
ORDER BY total_streams DESC;
```

### Most Popular Genre
Which music genre is the most popular?

```python
%sql
SELECT *
FROM gold_genre_popularity
ORDER BY total_streams DESC;
```

### Streams by Country
Which countries have the highest streaming activity?

```python
%sql
SELECT *
FROM gold_country_streams
ORDER BY total_streams DESC;
```

### Top 3 Artists


```python
%sql
SELECT *
FROM gold_top_artists
ORDER BY total_streams DESC
LIMIT 3;
```

### Countries with More Than 900 Streams

```python
%sql
SELECT *
FROM gold_country_streams
WHERE total_streams > 900
ORDER BY total_streams DESC;
```
