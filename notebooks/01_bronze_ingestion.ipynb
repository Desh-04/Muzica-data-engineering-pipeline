## # Step 1: Read the Table

```python
df = spark.table("dbacademy.default.streaming_data")
```

## # Step 2: Inspect the **Data**

### View the schema

```python
df.printSchema()
```

### View the first 10 records

```python
df.show(10)
```

### Count the records

```python
df.count()
```

## Step 3: Data Quality Check

```python
from pyspark.sql.functions import col, sum, when

df.select([
    sum(when(col(c).isNull(), 1).otherwise(0)).alias(c)
    for c in df.columns
]).show()
```

## Step 4: Look for Duplicate Records

```python
print("Rows before:", df.count())

df = df.dropDuplicates()

print("Rows after:", df.count())
```
