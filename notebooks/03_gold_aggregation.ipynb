## Gold Table 1 – Top Artists
Which artists receive the most streams?

```python
# Read the Silver table
silver_df = spark.table("silver_streams")
```

```python
#Aggregate
from pyspark.sql.functions import count
#Groups all rows by artist.
gold_top_artists = (
    silver_df
    .groupBy("artist")
    .agg(count("*").alias("total_streams"))
    .orderBy("total_streams", ascending=False)
)
```

**Save the table:**

```python
gold_top_artists.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("gold_top_artists")
```

## Gold Table 2 – Genre Popularity
Which music genre is most popular?

```python
gold_genre_popularity = (
    silver_df
    .groupBy("genre")
    .agg(count("*").alias("total_streams"))
    .orderBy("total_streams", ascending=False)
)
```

```python
gold_genre_popularity.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("gold_genre_popularity")
```

## Gold Table 3 – Country Streams
Which country generates the most streams?

```python
gold_country_streams = (
    silver_df
    .groupBy("country")
    .agg(count("*").alias("total_streams"))
    .orderBy("total_streams", ascending=False)
)
```

```python
gold_country_streams.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("gold_country_streams")
```
