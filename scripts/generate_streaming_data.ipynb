# Creating Sample Data for Muzica project

## 1. Import Libraries

```python
import pandas as pd
import random
from datetime import datetime, timedelta
```

## 2. Create Sample Data

```python
songs = [
    "Blinding Lights",
    "Believer",
    "Shape of You",
    "Levitating",
    "Perfect",
    "Thunder",
    "Stay",
    "Bad Habits",
    "Senorita",
    "Peaches"
]

artists = [
    "The Weeknd",
    "Imagine Dragons",
    "Ed Sheeran",
    "Dua Lipa",
    "Justin Bieber"
]

genres = [
    "Pop",
    "Rock",
    "Hip Hop",
    "POP",
    "rock",
    None
]

countries = [
    "India",
    "USA",
    "UK",
    "Canada",
    "Australia"
]

devices = [
    "Mobile",
    "Desktop",
    "Tablet"
]
```

## 3.Generate Records

```python
streaming_data = []
```

```python
start_date = datetime(2026, 7, 1)

for i in range(1, 5001):  #This creates 5,000 streaming events.

    record = {
        "stream_id": i,
        "user_id": random.randint(1000, 1500), #Generates random user IDs.
        "song_id": random.randint(1, 10),
        "song_name": random.choice(songs), #Randomly selects a song.
        "artist": random.choice(artists),
        "genre": random.choice(genres),
        "duration_seconds": random.randint(150, 300),
        "country": random.choice(countries),
        "device": random.choice(devices),
        "stream_timestamp": start_date + timedelta(
            minutes=random.randint(0, 10080) #There are 10,080 minutes in 7 days.
        )
    }

    streaming_data.append(record)
```

## 4.Convert to a DataFrame

```python
df = pd.DataFrame(streaming_data)
```
